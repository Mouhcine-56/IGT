```python
#d = 10, 50
def compute_error_metrics(v_exact, v_predicted, eps=1e-8):
    """
    Computes multiple error metrics between predicted and exact solution.
    Returns a dictionary with relative L2, relative Linf, MSE, and max relative abs error.
    """
    v_exact = v_exact.view(-1)
    v_predicted = v_predicted.view(-1)
    diff = v_predicted - v_exact

    rel_L2 = torch.norm(diff) / torch.norm(v_exact)
    rel_Linf = torch.norm(diff, p=float('inf'))
    mse = torch.mean(diff ** 2)


    return {
        'rel_L2': rel_L2.item(),
        'rel_Linf': rel_Linf.item(),
        'MSE': mse.item(),
    }

def test_metrics(V_NN, num_points, dim):
    x = 2 * torch.rand(num_points, dim, device=device) - 1  # Sample from [-1, 1]^dim
    #x = torch.linspace(-1, 1, num_points, device=device).unsqueeze(1)

    time_steps = [0.0, 0.25, 0.5, 0.75, 1.0]

    with torch.no_grad():
        V_NN.eval()
        for t_val in time_steps:
            t = torch.full((num_points, 1), t_val, device=device)
            v_pred = V_NN(t, x)
            v_exact = an.V_exact(x, t=t_val)

            metrics = compute_error_metrics(v_exact, v_pred)

            print(f"\n--- Time t = {t_val:.2f} ---")
            print(f"Relative L2 Error     : {metrics['rel_L2']:.4e}")
            print(f"Relative Linf Error   : {metrics['rel_Linf']:.4e}")
            print(f"MSE                   : {metrics['MSE']:.4e}")

# Example usage
test_metrics(V_NN, num_points=100000, dim=50)
```

```python
# d = 2
def create_meshgrid_2d(num_points):
    # Create 1D tensors for x and y coordinates
    x = torch.linspace(-1, 1, num_points, device=device)
    y = torch.linspace(-1, 1, num_points, device=device)

    # Create a meshgrid using torch.meshgrid()
    X, Y = torch.meshgrid(x, y)

    # Flatten X and Y and concatenate them
    X_flat = X.flatten()
    Y_flat = Y.flatten()
    points = torch.stack((X_flat, Y_flat), dim=1)

    return points

def compute_error_metrics(v_exact, v_predicted, eps=1e-8):
    """
    Computes multiple error metrics between predicted and exact solution.
    Returns a dictionary with relative L2, relative Linf, MSE, and max relative abs error.
    """
    v_exact = v_exact.view(-1)
    v_predicted = v_predicted.view(-1)
    diff = v_predicted - v_exact

    rel_L2 = torch.norm(diff) / torch.norm(v_exact)
    rel_Linf = torch.norm(diff, p=float('inf')) 
    mse = torch.mean(diff ** 2)
    max_rel_abs = torch.max(torch.abs(diff) / (torch.abs(v_exact) + eps))

    return {
        'rel_L2': rel_L2.item(),
        'rel_Linf': rel_Linf.item(),
        'MSE': mse.item(),

    }

def test_metrics(V_NN, num_points, dim):
    #x = 2 * torch.rand(num_points, dim, device=device) - 1  # Sample from [-1, 1]^dim
    #x = torch.linspace(-1, 1, num_points, device=device).unsqueeze(1)
    x = create_meshgrid_2d(num_points)

    time_steps = [0.0, 0.25, 0.5, 0.75, 1.0]

    with torch.no_grad():
        V_NN.eval()
        for t_val in time_steps:
            t = torch.full((num_points*num_points, 1), t_val, device=device)
            v_pred = V_NN(t, x)
            v_exact = an.V_exact(x, t=t_val)

            metrics = compute_error_metrics(v_exact, v_pred)

            print(f"\n--- Time t = {t_val:.2f} ---")
            print(f"Relative L2 Error     : {metrics['rel_L2']:.4e}")
            print(f"Relative Linf Error   : {metrics['rel_Linf']:.4e}")
            print(f"MSE                   : {metrics['MSE']:.4e}")

# Example usage
test_metrics(V_NN, num_points=100, dim=2)
```

```python
# control  d = 100
import math

def u_exact(x, t):
    denom = 1 + math.exp(2*(t - 1))
    return (-2 * x / denom).view(-1, 1)



def u_NN_from_V(PhiNN, x, t_val):
    """
    Compute u = -(1/2) * dPhi/dx using the internal analytic gradient getGrad().
    """
    # Construire [x ; t]
    t = torch.full((x.shape[0], 1), t_val, dtype=x.dtype, device=x.device)


    # Gradient complet wrt (x, t)
    grad_xt = PhiNN.get_grad(t, x)        # shape: [N, d+1]


    # Contrle optimal
    u_nn = -0.5 * grad_xt

    return u_nn

def test_control_metrics(PhiNN, num_points=100000, dim=100):
    x = 2 * torch.rand(num_points, dim) - 1
    time_steps = [0.0, 0.25, 0.5, 0.75, 1.0]

    PhiNN.eval()

    for t_val in time_steps:

        u_pred = u_NN_from_V(PhiNN, x, t_val)
        u_ex = u_exact(x, t_val)

        metrics = compute_error_metrics(u_ex, u_pred)

        print(f"\n--- CONTROL at Time t = {t_val:.2f} ---")
        print(f"Relative L2 Error    : {metrics['rel_L2']:.4e}")
        print(f"Relative Linf Error  : {metrics['rel_Linf']:.4e}")


            
test_control_metrics(V_NN)
```
